```python
import numpy as np
import pandas as pd
from channel import ChannelParams, gvf_from_downstream
from tools import generate_bed_profile_from_downstream, normal_depth_rectangular
from gvf import determine_gvf_zone_xL_first
import matplotlib.pyplot as plt
import holoviews as hv
import seaborn as sns
hv.extension("bokeh")
```

```python
# Create a 200 km prism channel
L = 200_000.0   # m
dx = 50.0       # m, computational step
x = np.arange(0.0, L + dx, dx) # create channel distances
b = 80.0            # m, channel width
n = 0.018            # Manning n
```

```python
# set the downstream bed elevation
z_ds = 440 # m

# segment x into 3 subsections, each section has different slopes
# we also add some undulations A * np.sin(2.0 * np.pi * x_up / wavelength)
x_segments = [x[x <= 50_000], x[(x > 50_000) & (x <= 70_000)], x[(x > 70_000) & (x <= 200_000)]]
S_mean_segments = [5e-4, 2e-4, 4e-4] # slope, m/m
A_segments = [0.4, 0.4, 0.4] # undulation magnitude
wavelength_segments = [50_000, 50_000, 50_000] # undulation wavelength
zb_segments = []
# we generate bed profile from downstream to upstream
# initialize the downstream bed elevation at the channel downstream
z_ds_seg = z_ds
for x_seg, S_mean, A, wavelength in zip(x_segments[::-1], S_mean_segments[::-1], A_segments[::-1], wavelength_segments[::-1]):
  # zb of this segement
  zb_segment = generate_bed_profile_from_downstream(x_seg, z_ds_seg, S_mean, A, wavelength)
  # update the downstream bed elevation for each segement
  z_ds_seg = zb_segment[0]
  # insert zb_segment to the first one
  zb_segments.insert(0, zb_segment)
zb = np.concatenate(zb_segments)
# quick plot the zb
plt.plot(x, zb)
```

```python
# Quick test of the normal depth given a q condition
yn_ups = normal_depth_rectangular(3000, b, n, S_mean_segments[0]) # 3000 m3/s
yn_ds = normal_depth_rectangular(3000, b, n, S_mean_segments[-1]) # 3000 m3/s
# add normal depth to the last point of zb
print(f"bed elevation at the downstream control is: {z_ds}")
print(f"normal depth elevation at the downstream: {z_ds + yn_ds}")
print(f"upstream normal depth: {zb[0] + yn_ups}")
```

```python
# Define scenarios with different Q (discharge) and different downstream water surface elevation
Q_list = [1000, 2000, 3000] # m3/s
wse_ds_list = [460, 490, 500] # m3/s

# create a scenario list based on these conditions
scenarios = {
    f"Q{Q}_wse{wse_ds:g}": {"Q": float(Q), "wse_ds": float(wse_ds)}
    for Q in Q_list
    for wse_ds in wse_ds_list
}

print("number of scenarios: ", len(scenarios))
```

```python
# Start simulation

results = {}  # results[name] = {"wse": ..., "y_n_elev": ..., "Q": ..., "wse_ds": ...}

for name, sc in scenarios.items():
    Q = sc["Q"]
    wse_ds = sc["wse_ds"]
    # normal depth elevation in different segments
    yn_elevation = []
    for zb_seg, S_mean in zip(zb_segments, S_mean_segments):
        normal_depth = normal_depth_rectangular(Q, b, n, S_mean)
        yn_seg = zb_seg + normal_depth
        yn_elevation.append(yn_seg)
    yn_elevation = np.concatenate(yn_elevation)

    # routing
    p = ChannelParams(b=b, n=n, Q=Q)
    y, wse = gvf_from_downstream(x, zb, wse_ds, p, method="rk4")

    results[name] = {
        "Q": Q,
        "wse_ds": wse_ds,
        "wse": wse,
        "yn_elevation": yn_elevation,
        "y": y,
    }
```

# Determine the xTr and xL and quick plotting the results

```python
def identify_xTr_xL_and_plot(
    x,
    zb,
    results,
    offset_thresh_m = 0.5, # m, threshold between wse and wse_normal to determine xTr
    slope_thresh_m_per_km = 0.01, # m/km, slope threshold for xL determination
    min_run_L = 10, # determine xL persistently for min_run_L steps, 10 steps means 500 m
    min_run_tr = 10, # wse persistently greater than wse_normal for next min_run_tr steps.
    smooth_window=9,
    x_units="km",
    cmap="tab10",
):
    # x-axis handling
    if x_units == "km":
        x_plot = x / 1000.0
        x_col = "x"
        x_axis_title = "Distance [km]"
        x_scale = 1000.0
    else:
        x_plot = x
        x_col = "x"
        x_axis_title = "Distance [m]"
        x_scale = 1.0

    # obtain scenario names
    scenario_names = list(results.keys())
    palette = sns.color_palette(cmap, len(scenario_names))
    scenario_colors = {name: palette[i] for i, name in enumerate(scenario_names)}
    scenario_overlay_dict = {}

    # loop each scenario
    for scenario_name, res in results.items():
        color = scenario_colors[scenario_name]
        # obtain simulated wse and wse_normal
        wse = np.asarray(res["wse"])
        wse_n = np.asarray(res["yn_elevation"])

        gvf_results = determine_gvf_zone_xL_first(
            x=x,
            wse=wse,
            wse_normal=wse_n,
            offset_thresh_m=offset_thresh_m,
            slope_thresh_m_per_km=slope_thresh_m_per_km,
            min_run_tr=min_run_tr,
            min_run_L=min_run_L,
            smooth_window=smooth_window,
        )

        df = pd.DataFrame({
            x_col: x_plot,
            "wse": wse,
            "wse_normal": wse_n,
            "scenario": scenario_name,
            "offset": gvf_results.offset,
            "slope_m_per_km": gvf_results.slope_m_per_m * 1000.0,
        })

        # GVF curve with hover_tooltips
        wse_curve = hv.Curve(df, kdims=[x_col], vdims=["wse"]).opts(
            color=color,
            line_width=2,
            tools=["hover"],
            hover_tooltips=[
                (x_axis_title, f"@{x_col}{{0.00}}"),
                ("WSE [m]", "@wse{0.000}"),
            ],
        )

        # Normal curve with hover_tooltips
        normal_curve = hv.Curve(df, kdims=[x_col], vdims=["wse_normal"]).opts(
            color=color,
            line_dash="dashed",
            alpha=0.8,
            tools=["hover"],
            hover_tooltips=[
                (x_axis_title, f"@{x_col}{{0.00}}"),
                ("WSE_normal [m]", "@wse_normal{0.000}"),
            ],
        )

        layers = [wse_curve, normal_curve]

        # Marker DataFrames (Points need x,y as columns)
        if gvf_results.idx_tr is not None:
            i = gvf_results.idx_tr
            mtr = pd.DataFrame([{
                x_col: x_plot[i],
                "y": wse[i],
                "scenario": scenario_name,
                "kind": "xTr",
                "offset": gvf_results.offset[i],
                "slope_m_per_km": gvf_results.slope_m_per_m[i] * 1000.0,
            }])
            layers.append(
                hv.Points(mtr, kdims=[x_col, "y"], vdims=["scenario", ]).opts(
                    color=color,
                    marker="triangle",
                    size=12,
                    tools=["hover"],
                )
            )

        if gvf_results.idx_L is not None:
            i = gvf_results.idx_L
            mL = pd.DataFrame([{
                x_col: x_plot[i],
                "y": wse[i],
                "scenario": scenario_name,
                "kind": "xL",
                "offset": gvf_results.offset[i],
                "slope_m_per_km": gvf_results.slope_m_per_m[i] * 1000.0,
            }])
            layers.append(
                hv.Points(mL, kdims=[x_col, "y"], vdims=["scenario",]).opts(
                    color=color,
                    marker="inverted_triangle",
                    size=12,
                    tools=["hover"],
                )
            )

        scenario_overlay_dict[scenario_name] = hv.Overlay(layers)

    scenarios = hv.NdOverlay(scenario_overlay_dict, kdims=["Scenario"]).opts(
        legend_position="right",
        click_policy="hide",
    )

    bed_df = pd.DataFrame({x_col: x_plot, "zb": zb})
    bed = hv.Curve(bed_df, kdims=[x_col], vdims=["zb"], label="Bed profile").opts(
        color="black",
        line_width=2,
        tools=["hover"],
        hover_tooltips=[
            (x_axis_title, f"@{x_col}{{0.00}}"),
            ("zb [m]", "@zb{0.000}"),
        ],
    )

    return (bed * scenarios).opts(
        width=1000,
        height=600,
        title="GVF profiles (one legend entry per scenario)",
        xlabel=x_axis_title,
        ylabel="Elevation [m]",
        tools=["pan", "wheel_zoom", "box_zoom", "reset", "save", "hover"],
        show_grid=True,
    )
```

```python
# Plot results
hv_plot = identify_xTr_xL_and_plot(
    x=x,
    zb=zb,
    results=results,
    offset_thresh_m=0.5,
    slope_thresh_m_per_km=0.01,
    x_units="km",
    cmap="Set1",  # or "Category20", "Set1"
)

hv_plot
```
