```python
import numpy as np
import pandas as pd
from channel import ChannelParams, gvf_from_downstream
from tools import generate_bed_profile_from_downstream, normal_depth_rectangular
from gvf import determine_gvf_zone_xL_first
import matplotlib.pyplot as plt
# from dataclasses import dataclass
# from typing import Literal, Tuple, Optional
# import holoviews as hv
# import seaborn as sns
# from itertools import cycle
# hv.extension("bokeh")
```

```python
# Create a 200 km prism channel
L = 200_000.0   # m
dx = 50.0       # m, computational step
x = np.arange(0.0, L + dx, dx) # create channel distances
b = 80.0            # m, channel width
n = 0.018            # Manning n
```

```python
# Generate bed profiles with small undulations
# here I am showing how to generate a bed profile with different slopes
z_ds = 440 # bed elevation at the downstream, which is x[-1]
# segment x into 3 subsections
x_seg1 = x[x <= 100_000]
S_mean_1 = 2e-4 # m/m, 0.5m/km
x_seg2 = x[(x > 100_000) & (x <= 150_000)]
S_mean_2 = 1e-4 # flat for a while, 0.1m/km
x_seg3 = x[(x > 150_000) & (x <= 200_000)]
S_mean_3 = 2e-4

# generate bed profile from downstream
zb_seg3 = generate_bed_profile_from_downstream(x_seg3, z_ds, S_mean_3, A=0.4, wavelength=50_000)
zb_seg2 = generate_bed_profile_from_downstream(x_seg2, zb_seg3[0], S_mean_2, A=0.4, wavelength=50_000)
zb_seg1 = generate_bed_profile_from_downstream(x_seg1, zb_seg2[0], S_mean_1, A=0.4, wavelength=50_000)
zb = np.concatenate((zb_seg1, zb_seg2, zb_seg3))

# quick plot the zb
plt.plot(x_seg1, zb_seg1, color = "r")
plt.plot(x_seg2, zb_seg2, color = "g")
plt.plot(x_seg3, zb_seg3, color = "b")
```

```python
# Now we do a quick test to see what normal depth would be given a Q condition
def get_yn_downstream(Q):
    # normal depth elevation (3 segments)
    normal_depth_1 = normal_depth_rectangular(Q, b, n, S_mean_1)
    normal_depth_2 = normal_depth_rectangular(Q, b, n, S_mean_2)
    normal_depth_3 = normal_depth_rectangular(Q, b, n, S_mean_3)
    y_n_elevation = np.concatenate(
        (zb_seg1 + normal_depth_1, zb_seg2 + normal_depth_2, zb_seg3 + normal_depth_3)
    )
    return y_n_elevation[-1]

print("bed elevation at the downstream control is: ", z_ds)
print("normal depth elevation at the downstream: ", get_yn_downstream(3000)) # 3000 m3/s
```

This indicate we need to select the downstream water surface elevation at least to be 440m to make sure the program is running

- wse_ds > 451 -> M1 curve
- wse_ds < 451 -> M2 curve

```python
# Define scenarios with different Q (discharge) and different downstream water surface elevation
Q_list = [500, 800, 1000] # m3/s
wse_ds_list = [480, 490, 500] # m3/s

# create a scenario list based on these conditions
scenarios = {
    f"Q{Q}_wse{wse_ds:g}": {"Q": float(Q), "wse_ds": float(wse_ds)}
    for Q in Q_list
    for wse_ds in wse_ds_list
}

print("number of scenarios: ", len(scenarios))
```

```python
# Start simulation

results = {}  # results[name] = {"wse": ..., "y_n_elev": ..., "Q": ..., "wse_ds": ...}

for name, sc in scenarios.items():
    Q = sc["Q"]
    wse_ds = sc["wse_ds"]
    # normal depth elevation (3 segments)
    normal_depth_1 = normal_depth_rectangular(Q, b, n, S_mean_1)
    normal_depth_2 = normal_depth_rectangular(Q, b, n, S_mean_2)
    normal_depth_3 = normal_depth_rectangular(Q, b, n, S_mean_3)
    y_n_elevation = np.concatenate(
        (zb_seg1 + normal_depth_1, zb_seg2 + normal_depth_2, zb_seg3 + normal_depth_3)
    )
    # routing
    p = ChannelParams(b=b, n=n, Q=Q)
    y, wse = gvf_from_downstream(x, zb, wse_ds, p, method="rk4")

    results[name] = {
        "Q": Q,
        "wse_ds": wse_ds,
        "wse": wse,
        "y_n_elevation": y_n_elevation,
        "y": y,
    }
```
